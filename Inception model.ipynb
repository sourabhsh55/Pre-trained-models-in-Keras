```python
# importing libraries.
import os
import keras
import numpy as np
import matplotlib.pyplot as plt
from keras import preprocessing
from keras.applications import inception_v3
```

```python
# inception model trained on imagenet dataset.
model = inception_v3.InceptionV3(weights='imagenet')
```

```python
# inception model accept input image of dimensions: (299,299,3)
x = preprocessing.image.load_img('computer.jpg',target_size=(299,299))
plt.imshow(x)
plt.show()
# image in array form.
x = preprocessing.image.img_to_array(x)
# expanding dimensions of the image: (1,299,299,3)
x = np.expand_dims(x,axis=0)
x = inception_v3.preprocess_input(x)
predictions = model.predict(x) # predictions.
labels = inception_v3.decode_predictions(predictions,top=3)[0] # top 3 results acc to the model.
labels
```
